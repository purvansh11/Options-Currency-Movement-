```python
import pandas as pd
dataset = "Foreign_Exchange_Rates.csv"
df = pd.read_csv(dataset)
df.head()
```

```python
df.info()
```

```python
updated_df_without_nd = df[df.columns.values[2]]!="ND" 
df = df[updated_df_without_nd]
df.dropna(axis=0,inplace=True)
```

```python
cols = []
ind = 0
col_head = {}
regions = []
for col in df.columns:
    split_name = col.split()
    print(split_name[0])
    regions.append(split_name[0])
#     region = split_name[0]

for region in regions:
    col_head[region] = df.columns.values[ind]
    ind += 1
```

```python
election_dates = ['2000-09-07' , '2004-09-02' , '2008-09-04' , '2012-09-06' , '2016-09-08']
year_to_date = {}
for date in election_dates:
    year_to_date[date[:4]] = date
```

```python
from tabulate import tabulate
cost = {}
hash_value_dataframe = {}
hash_value_dates = {}
dates = []
rows = df.shape[0]
ind = 0
for i in range(rows+1):
    try:
        dates.append(df[col_head["Time"]][i])
        hash_value_dataframe[df[col_head["Time"]][i]] = i
        hash_value_dates[df[col_head["Time"]][i]] = ind
        ind += 1
    except:
        pass

for date in dates:
    cost[date] = {}
for date in dates:
    for region in regions:
        try:
            cost[date][region] = float(df[col_head[region]][hash_value_dataframe[date]]) 
        except:
            pass
# for _date in dates:
#     for region in regions:
#         try:
#             cost[_date][region] = float(cost[_date][region])
#         except:
#             continue
print("Cost on 2004-09-02 in THAILAND is: ",end='')
print(cost['2004-09-02']["THAILAND"])

results = []
buffer = 60
for date in election_dates:
    start_date = dates[hash_value_dates[date]-buffer]
    end_date = dates[hash_value_dates[date]+buffer]
    start_index = hash_value_dates[start_date]
    end_index = hash_value_dates[end_date]

    for region in regions:
        if region == "Unnamed:" or region == "Time":
            continue
        
        before_60 = cost[dates[start_index]][region]
        on_day = cost[date][region]
        after_60 = cost[dates[end_index]][region]
        
        normalize_before = round((float)(before_60) - (float)(on_day),10)
        normalize_after = round((float)(after_60) - (float)(on_day),10)
        
        results.append((region,before_60,on_day,after_60,normalize_before,normalize_after))
    results.append(("-------","-------","-------","-------","-------","-------"))

# print(tabulate(results,headers = ["Region","Before 60 Days","On Election Day","After 60 Days","Normalize Before","Normalize After"]))
```

```python
currency_of = {}
region_of = {}
currency_of["AUSTRALIA"] = "AUD"
currency_of["EURO"] = "EUR"
currency_of["NEW"] = "NZD"
currency_of["UNITED"] = "GBP"
currency_of["BRAZIL"] = "BRL"
currency_of["CANADA"] = "CAD"
currency_of["CHINA"] = "CNY"
currency_of["HONG"] = "HKD"
currency_of["INDIA"] = "INR"
currency_of["KOREA"] = "KRW"
currency_of["MEXICO"] = "MXN"
currency_of["SOUTH"] = "ZAR"
currency_of["SINGAPORE"] = "SGD"
currency_of["DENMARK"] = "DKK"
currency_of["JAPAN"] = "JPY"
currency_of["MALAYSIA"] = "MYR"
currency_of["NORWAY"] = "NOK"
currency_of["SWEDEN"] = "SEK"
currency_of["SRI"] = "LKR"
currency_of["SWITZERLAND"] = "CHF"
currency_of["TAIWAN"] = "TWD"
currency_of["THAILAND"] = "THB"
for region in regions:
    if region == "Unnamed:" or region == "Time":
        continue
    region_of[currency_of[region]] = region 
```

```python
import matplotlib.pyplot as plt
traded_at_2d = {}

def traded_plot(currency,timespan):
    buffer_plot = 120
    mini = 1000
    maxi = -1000
    for date in election_dates:
        traded_at_2d[date] = []
    for date in election_dates:
        plot_start_index = hash_value_dates[date]-buffer_plot
        plot_start_date = dates[plot_start_index]
        plot_end_index = hash_value_dates[date]+buffer_plot
        plot_end_date = dates[plot_end_index]
        traded_at = []
        x_values = []
        x_val = -buffer_plot

        for i in range(plot_start_index,plot_end_index+1):
            try:
                traded_at_2d[date].append(round((float(cost[dates[i]][region_of[currency]])),10))
                traded_at.append(round(float(cost[dates[i]][region_of[currency]]) - float(cost[date][region_of[currency]]),10))
                x_values.append(x_val)
                x_val += 1
            except Exception as e:
                continue
        plt.plot(x_values,traded_at)
        mini = min(mini,min(traded_at))
        maxi = max(maxi,max(traded_at))
#         traded_at_2d[date] = traded_at
    # aus_plot.axes.get_yaxis().set_visible(False)
    plt.axvline(x=-timespan*30, color='black', linestyle='--')
    plt.axvline(x=timespan*30, color='black', linestyle='--')
    plt.axhline(y=mini, color='red', linestyle='--')
    plt.axhline(y=maxi, color='green', linestyle='--')
    plt.title("Traded plot for {} currency".format(currency))
    plt.grid()
    plt.show()
    ind=0
    # for i in range(-buffer_plot,buffer_plot+1):
    #     print(i,traded_at[i+buffer_plot],ind)
    #     ind+=1
traded_plot("NOK",2)
```
